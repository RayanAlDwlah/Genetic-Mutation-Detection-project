# Phase 0: Data Download & Verification

This notebook documents the raw data sources used in our project,
their download links, file sizes, and integrity verification.


```python
# Data Sources Summary
import pandas as pd
import os
from datetime import datetime

sources = {
    'Dataset': ['ClinVar', 'gnomAD', 'dbNSFP', 'UniProt'],
    'Description': [
        'Clinically classified genetic variants',
        'Population allele frequencies (140K+ individuals)',
        'Pre-computed functional features for missense variants',
        'Human protein sequences (for ESM-2 model)'
    ],
    'Version': ['2026-02', 'r2.1.1', '5.3.1a', '2025_01'],
    'Genome Build': ['GRCh37', 'GRCh37', 'GRCh37', 'N/A'],
    'Source': ['NCBI', 'Broad Institute', 'jpopgen', 'UniProt.org']
}

df_sources = pd.DataFrame(sources)
print('=' * 70)
print('DATA SOURCES USED IN THIS PROJECT')
print('=' * 70)
print(df_sources.to_string(index=False))
print()
```

```python
# Verify raw files exist and show sizes
import os

raw_dir = '../data/raw'

files_to_check = [
    ('clinvar/variant_summary.txt.gz',              'ClinVar variants'),
    ('gnomad/gnomad.exomes.r2.1.1.sites.vcf.bgz',  'gnomAD exome frequencies'),
    ('dbnsfp/dbNSFP5.3.1a_grch37.gz',              'dbNSFP features'),
]

print('=' * 70)
print('RAW FILE VERIFICATION')
print('=' * 70)
print(f"{'File':<50} {'Size':>10} {'Status':>8}")
print('-' * 70)

all_present = True
for filepath, description in files_to_check:
    full_path = os.path.join(raw_dir, filepath)
    if os.path.exists(full_path):
        size_mb = os.path.getsize(full_path) / (1024 * 1024)
        size_str = f"{size_mb/1024:.1f} GB" if size_mb > 1024 else f"{size_mb:.1f} MB"
        print(f"{description:<50} {size_str:>10} {'OK':>8}")
    else:
        print(f"{description:<50} {'---':>10} {'MISSING':>8}")
        all_present = False

print('-' * 70)
if all_present:
    print('All raw files are present.')
else:
    print('WARNING: One or more files are missing. Download them before running the pipeline.')
```
